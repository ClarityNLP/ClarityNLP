```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime
```

```python
samples = pd.read_csv('data/nct_features_sampled.csv')
samples = samples.set_index(samples.columns[0])
samples.index = range(samples.shape[0])
samples.index.name = 'id'
samples['match_count'] = samples.sum(axis=1)
samples['gender_count'] = samples['is_male'] + samples['is_female']
samples['gender_count'] = samples['gender_count'].apply(lambda x: 0.0 if x == 1.0 else x)
samples['final_feature_count'] = samples['match_count'] - samples['gender_count']
samples['logical_operations'] = samples['final_feature_count'].apply(lambda x: 1.0 if x >= 2.0 else 0.0)
samples
```

```python
start_id = 684
status = pd.read_csv('data/job_status.csv', names=['status_id', 'status', 'description', 'time', 'id'])
status['clarity_id'] = status['id']
status['id'] = status['id'] - start_id
status['time'] = pd.to_datetime(status['time'])
status.head(10)
```

```python
job_stats = pd.DataFrame(status.groupby('id')['time'].agg(np.ptp))
job_stats['seconds'] = job_stats['time'].apply(lambda x:  x.total_seconds())
job_stats['clarity_id'] = (job_stats.index + start_id)
```

```python
job_stats = pd.concat([job_stats, samples], axis=1).dropna()
job_stats
```

```python
average_time = np.mean(job_stats['seconds'])
average_time
```
